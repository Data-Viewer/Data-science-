### Subplot

    Subplot means a group of smaller axes Subplot divided the curent figure into rectangular panes that are numbered row-Wise

```python
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,1)
plt.plot(x,y,'o')

x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y,'o')


plt.show()
```

### Bar plot

    A bar plot is a plot that presents categorical data with rectangular bars with length proportional to the values that they represent  

```python
lags=["c","c++","Java","python"]
student=[1,2,3,4]
plt.bar(lags,student,color='g')
plt.show()
```

### Histogram

    A histogram is a graphical representation of data. It is used to summarize discreat or continuous data that are measured on an interval scale

```python
fig,ax=plt.subplots(1,1)
a=np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
ax.hist(a,bins=[0,25,50,75,100])
ax.set_title("Histogram")
ax.set_xlabel("marks")
ax.set_ylabel("no. of students")
plt.show()
```

### Pie plot

    A pie plot is a proportional representation of the numerical data in a column 

```python
nums=[123,234,456,345,333]
labels=['india','pakistan','china','north','east']
explode=[0,0,0.2,0,0]
colors=['black','pink','blue','g','y']
plt.pie(nums,labels=labels,explode=explode, shadow=True,startangle=90,colors=colors)
plt.title('population')
plt.grid(True)
plt.show()
```

### Scatterplot

```python
x=[0,1,2,3]
y=[3,4,5,6]
x1=[6,7,8,9]
y1=[2,3,4,4]
plt.scatter (x,y,label='High income', color='r')
plt.scatter (x1,y1,label='Low income',color='b')
plt.ylabel('saving')
plt.xlabel('income')
plt.title('scatter')
plt.legend()
plt.show()
```

```python
girl_grades=[80,20,30,40]
boy_grades=[20,30,40,50]
grades_range=[10,20,40,100]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(grades_range,girl_grades,color='r')
ax.scatter(grades_range,boy_grades,color='b')
ax.set_xlabel('Grades')
ax.set_ylabel('scores')
ax.set_title('scatter plot')
plt.legend()
plt.show()
```
